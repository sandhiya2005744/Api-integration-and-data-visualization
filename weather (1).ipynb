```python
import pandas as pd

# Sample data
weather_data = {
    "City": ["Chennai", "Delhi", "Mumbai"],
    "Temperature (C)": [32, 35, 30],
    "Humidity (%)": [70, 50, 65]
}

df = pd.DataFrame(weather_data)
print(df)

# Plot
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")
plt.figure(figsize=(8, 5))
sns.barplot(x="City", y="Temperature (C)", data=df, palette="coolwarm")
plt.title("Temperature in Different Cities")
plt.ylabel("Temperature (C)")
plt.xlabel("City")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.lineplot(x="City", y="Humidity (%)", data=df, color="green")
plt.title("Humidity in Different Cities")
plt.ylabel("Humidity (%)")
plt.xlabel("City")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Sample DataFrame
weather_data = {
    "City": ["Chennai", "Delhi", "Mumbai", "Kolkata", "Bangalore"],
    "Pressure (hPa)": [1012, 1015, 1010, 1008, 1013]
}

df = pd.DataFrame(weather_data)

# Plot
sns.set(style="whitegrid")
plt.figure(figsize=(8, 5))
sns.barplot(x="City", y="Pressure (hPa)", data=df, palette="Blues_d")
plt.title(" Atmospheric Pressure in Different Cities")
plt.ylabel("Pressure (hPa)")
plt.xlabel("City")
plt.show()
```
